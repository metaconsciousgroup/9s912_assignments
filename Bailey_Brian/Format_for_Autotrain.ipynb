```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np
```

```python
# load data from parquet

num_conversation_chunk_files=5
last_i_not_found = 0

for i in range(1, num_conversation_chunk_files):
    try:
        globals()[f'conversations_df{i}'] = pd.read_parquet(f'/content/drive/My Drive/The Football Archive™/Training Data/initial-training-data-{i}.parquet')
    except FileNotFoundError:
        if not i == last_i_not_found + 1:
            print(f"{i} not found")
        last_i_not_found=i
        globals()[f'conversations_df{i}'] = pd.DataFrame()

conversations_df = pd.concat([globals()[f'conversations_df{i}'] for i in range(1, num_conversation_chunk_files)], axis=0)
conversations_df
```

```python
conversations_df.reset_index(inplace=True)
```

```python
type(conversations_df['conversation'][0])
```

```python
import re

def replace_roles(text):
    # Replace "Player: " with "### Human:"
    text = re.sub(r'Player: ', '### Human: ', text)

    # Replace "Coach: " with "### Assistant:"
    text = re.sub(r'Coach: ', '### Assistant: ', text)

    text = text.replace("\n", " ")

    text = "### System: You are an AI Assistant with deep knowledge of American football. " + text

    return text
```

```python
conversations_df['text'] = conversations_df['conversation'].apply(replace_roles)
conversations_df
```
